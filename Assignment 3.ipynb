```python
def sum(num):
    sum = 0
    for i in num:
        sum+=i
    return sum
list=(8,2,3,0,7)
print(sum(list))
```

```python
def reverse(s):
    str = " "
    for i in s:
        str = i+str
    return str
s = "1234abcd"
print(" ",end=" ")
print(reverse(s))
```

```python
def up_low(str):
    upper = 0
    lower = 0
    for char in str:
        if char.islower():
            lower += 1
        elif char.isupper():
            upper += 1
        else:
            pass
    return(upper,lower)
print(up_low("The quick Brow Fox"))
```
